Questions to be answered
1. What are the names of all the videos and their corresponding channels?

```python
video_channel=pd.DataFrame()
cursor.execute('select video_name,channel_name from Video')
records = cursor.fetchall()

# Print the records
for record in records:
    video_channel=video_channel.append({'Video_name':record[0],'Channel_name':record[1]},ignore_index=True)
video_channel
```

2.Which channels have the most number of videos, and how many videos do
 they have


```python
count_of_videos=pd.DataFrame()
cursor.execute('select count(video_name) as count_vids,channel_name from Video group by channel_name')
records = cursor.fetchall()
for record in records:
  count_of_videos=count_of_videos.append({'video_counts':record[0],'Channel_name':record[1]},ignore_index=True)
count_of_videos
```

3.What are the top 10 most viewed videos and their respective channels?

```python
max_viewed_videos=pd.DataFrame()
cursor.execute('select video_name,channel_name from Video group by video_name,channel_name order by max(Video_views) desc limit 10 ')
records = cursor.fetchall()
for record in records:
  max_viewed_videos=max_viewed_videos.append({'video_name':record[0],'channel_name':record[1]},ignore_index=True)
max_viewed_videos
```

4.How many comments were made on each video, and what are their
 corresponding video names?


```python
number_of_comments=pd.DataFrame()
cursor.execute('select Video.video_id,Video.video_name,count(Comments.comments_id) as comments_count from Video inner join Comments on Video.video_id=Comments.video_id group by Video.video_id,Video.video_name')
records = cursor.fetchall()
for record in records:
  number_of_comments=number_of_comments.append({'Video_id':record[0],'Video_name':record[1],'Comments_counts':record[2]},ignore_index=True)
number_of_comments
```

5.Which videos have the highest number of likes, and what are their
corresponding channel names?


```python
highes_likes_df=pd.DataFrame()
cursor.execute('select max(Video_Likes_count) as Max_likes,Video_name,Channel_name from Video group by Video_name,Channel_name order by max(Video_Likes_count) desc')
records = cursor.fetchall()
for record in records:
  highes_likes_df=highes_likes_df.append({'Likes':record[0],'Video_name':record[1],'Channel_name':record[2]},ignore_index=True)
highes_likes_df
```

6.What is the total number of likes and dislikes for each video, and what are
their corresponding video names?


```python
highes_likes_df=pd.DataFrame()
cursor.execute('select max(Video_Likes_count) as Max_likes,Video_name,Channel_name from Video group by Video_name,Channel_name order by max(Video_Likes_count) desc')
records = cursor.fetchall()
for record in records:
  highes_likes_df=highes_likes_df.append({'Likes':record[0],'Video_name':record[1],'Channel_name':record[2]},ignore_index=True)
highes_likes_df
```

```python
highes_likes_df2=highes_likes_df.groupby('Channel_name')['Likes'].sum()
highes_likes_df2=highes_likes_df2.reset_index(drop=False)
highes_likes_df2
```

7.What is the total number of views for each channel, and what are their
corresponding channel names?


```python
Views_channel=pd.DataFrame()
cursor.execute('select channel_name,channel_views from Channel')
records = cursor.fetchall()
for record in records:
  Views_channel=Views_channel.append({'Channel_name':record[0],'View':record[1]},ignore_index=True)
Views_channel
```

8.What are the names of all the channels that have published videos in the year
 2022


```python
#Video_published_date
video_pub_df=pd.DataFrame()
dat=str('2022-01-01T00:00:00z')
cursor.execute('select video_name,Video_published_date,Channel_name from Video where Video_published_date >='+"'{dat}'".format(dat=dat))
records = cursor.fetchall()
for record in records:
  video_pub_df=video_pub_df.append({'Video_name':record[0],'Published_date':record[1],'Channel_name':record[2]},ignore_index=True)
video_pub_df
```

```python
#highest comment
Highest_comment_df=pd.DataFrame()
cursor.execute('select Video.video_id,Video.video_name,(count(Comments.comments_id)) as comments_count,Video.channel_name from Video inner join Comments on Video.video_id=Comments.video_id \
group by Video.video_id,Video.video_name,Video.channel_name')
records = cursor.fetchall()
for record in records:
  Highest_comment_df=Highest_comment_df.append({'Channel_name':record[3],'Comments':record[2]},ignore_index=True)
Highest_comment_df
```

```python
Highest_comment_df=Highest_comment_df.groupby('Channel_name')['Comments'].sum()
Highest_comment_df
```
